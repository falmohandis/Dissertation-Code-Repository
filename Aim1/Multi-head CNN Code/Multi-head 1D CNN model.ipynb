## Data Leakage Control Algorithm

This code ensures that there is no overlap among indices allocated for the training, validation, and test sets.

1.   check_overlap Function:
*   This function takes three lists of indices (train_indices, validation_indices, test_indices) as arguments.
*   It first converts these lists into sets for efficient intersection checks.
*   Then it checks for intersection among these sets. If any overlap is found, it raises a ValueError indicating the presence of overlap.
*   If no overlap is found, it prints a message confirming that no overlap exists.

2.   Implementation:
*   The function check_overlap is invoked with the indices of the train, validation, and test sets as arguments.
*   If no overlap is detected, a message "Overlap check passed" is printed.
*   If an overlap is found, the ValueError message is printed, indicating the type of overlap detected (train-validation, train-test, or validation-test).
*   If an overlap is detected, the program exits after printing the error message.

```python
def check_overlap(train_indices, validation_indices, test_indices):
    """
    Function to check for overlap among indices of train, validation, and test sets.

    Args:
    train_indices (list): List of indices for the train set.
    validation_indices (list): List of indices for the validation set.
    test_indices (list): List of indices for the test set.

    Raises:
    ValueError: If overlap is found among indices.
    """
    # Convert indices to sets for efficient intersection check
    train_set = set(train_indices)
    validation_set = set(validation_indices)
    test_set = set(test_indices)

    # Check for intersection among sets
    if train_set.intersection(validation_set) or \
       train_set.intersection(test_set) or \
       validation_set.intersection(test_set):
        raise ValueError("Overlap found among train, validation, and test sets.")
    else:
        print("No overlap found among train, validation, and test sets.")

#  Implement:
try:
    check_overlap(train_indices, validation_indices, test_indices)
    print('Overlap check passed')
except ValueError as e:
    print(e)  # Print the error message
    # Handle the error or exit the program as needed
    # Exiting the program here for demonstration purposes
    exit()
```

# Class Balancer

```python
import numpy as np

def balance_classes(X, y):
    # Find the unique classes and their counts
    unique_classes, counts = np.unique(y, return_counts=True)

    # Determine the minimum count among the classes
    min_count = min(counts)

    # Create lists to store the balanced dataset
    X_balanced = []
    y_balanced = []

    # Loop through each class
    for cls in unique_classes:
        # Get the indices of all samples in this class
        cls_indices = np.where(y == cls)[0]

        # Randomly shuffle the indices and select min_count samples
        np.random.shuffle(cls_indices)
        selected_indices = cls_indices[:min_count]

        # Add the selected samples to the balanced dataset
        X_balanced.append(X[selected_indices])
        y_balanced.append(y[selected_indices])

    # Concatenate all the balanced class arrays into final balanced datasets
    X_balanced = np.concatenate(X_balanced)
    y_balanced = np.concatenate(y_balanced)

    return X_balanced, y_balanced

def print_class_distribution(y, dataset_name):
    unique_classes, counts = np.unique(y, return_counts=True)
    print(f"{dataset_name} class distribution:")
    for cls, count in zip(unique_classes, counts):
        print(f"Class {cls}: {count} samples")
    print()

# Apply the balance_classes function to the training, validation, and test sets
X_train_balanced, y_train_balanced = balance_classes(X_train, y_train)
X_valid_balanced, y_valid_balanced = balance_classes(X_valid, y_valid)
X_test_balanced, y_test_balanced = balance_classes(X_test, y_test)

# Print the shapes of the balanced datasets
print("Balanced Train set shape:", X_train_balanced.shape, y_train_balanced.shape)
print("Balanced Validation set shape:", X_valid_balanced.shape, y_valid_balanced.shape)
print("Balanced Test set shape:", X_test_balanced.shape, y_test_balanced.shape)

# Overwrite the original datasets with the balanced datasets
X_train, y_train = X_train_balanced, y_train_balanced
X_valid, y_valid = X_valid_balanced, y_valid_balanced
X_test, y_test = X_test_balanced, y_test_balanced

# Print the class distribution for the balanced training, validation, and test sets
print_class_distribution(y_train_balanced, "Balanced Train")
print_class_distribution(y_valid_balanced, "Balanced Validation")
print_class_distribution(y_test_balanced, "Balanced Test")



```

## Multi-Head CNN Architecture Definition
This part of the code defines the architecture for a multi-head convolutional neural network (CNN) using multiple convolutional layers for feature extraction from each channel of the input data, followed by flattening, concatenation, and dense layers for classification.

1. Channel Input Creation:
* It iterates through the channels of the input data (X), creating an input layer for each channel. Each input layer has the shape (X.shape[1], 1), where X.shape[1] represents the length of each sample.
2. Convolutional Layers:
* For each channel, it adds three convolutional layers (Conv1D) followed by max pooling (MaxPooling1D).
* Each convolutional layer has a specified number of filters and kernel size, with ReLU activation and L2 regularization.
* The output of each convolutional layer is pooled to reduce dimensionality.
3. Flattening and Concatenation:
* After the convolutional layers, the outputs are flattened and stored in a list.
* All the flattened outputs are concatenated into a single tensor.

## Information pertaining to L1 vs L2 normalization.


See the below:

 From a practical standpoint, L1 tends to shrink coefficients to zero whereas L2 tends to shrink coefficients evenly. L1 is therefore useful for feature selection, as we can drop any variables associated with coefficients that go to zero. L2, on the other hand, is useful when you have collinear/codependent features. (An example pair of codependent features is gender and ispregnant since, at the current level of medical technology, only females can be ispregnant.) Codependence tends to increase coefficient variance, making coefficients unreliable/unstable, which hurts model generality. L2 reduces the variance of these estimates, which counteracts the effect of codependencies.

### One-Hot Encoding in Detail:

* One-hot encoding is a process used to convert categorical data, such as class labels, into a numerical format suitable for machine learning algorithms. Here's how it works:

1. Categorical Labels: Suppose you have a set of labels, each representing a class or category, e.g., ["cat", "dog", "bird"].
2. Integer Encoding: First, these labels are encoded into integers. For example, "cat" might be encoded as 0, "dog" as 1, and "bird" as 2.
3. One-Hot Encoding: In one-hot encoding, each integer-encoded label is represented as a binary vector. This vector has the same length as the total number of classes, with a 1 in the position corresponding to the label's class and 0s elsewhere.
* So in our example: \
"cat" (encoded as 0) becomes [1, 0, 0]. \
"dog" (encoded as 1) becomes [0, 1, 0]. \
"bird" (encoded as 2) becomes [0, 0, 1].

* This encoding scheme ensures that each class is represented as a distinct and orthogonal vector, facilitating learning in machine learning models. It's commonly used in classification tasks with categorical labels.

## Interpreting the Above Outputs:

Loss and accuracy are metrics used to assess the performance of the model on both training and validation datasets, while validation loss and validation accuracy specifically focus on evaluating the model's generalization performance on unseen data.


1. Loss:
* The loss/cost function measures how well the model's predictions match the actual target values (or ground truth) in the training data. It quantifies the difference between predicted and true values.
* Lower loss values indicate better agreement between predictions and actual values.
2. Accuracy:
* Accuracy is a measure of the model's performance in correctly predicting the labels of the data points.
* It represents the proportion of correctly classified data points out of the total number of data points.
* It is a metric to evaluate how well the model is performing on the dataset.

3. Validation Loss:
* Validation loss is similar to training loss, but it is calculated on a separate validation dataset rather than the training dataset.
* Validation loss helps in assessing how well the model generalizes to new, unseen data.
* It indicates whether the model is overfitting or underfitting.
* Ideally, both training and validation loss should decrease during training, but if the validation loss starts to increase while the training loss decreases, it suggests overfitting.

4. Validation Accuracy:
*Validation accuracy is similar to accuracy but calculated on the validation dataset.
* It measures the proportion of correctly classified data points in the validation dataset.
* It provides insights into whether the model is learning meaningful patterns from the data or simply memorizing the training examples.

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import roc_auc_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

# Generate predictions on the test set
y_pred = model.predict([X_test[:, :, i].reshape((X_test.shape[0], X_test.shape[1], 1)) for i in range(X_test.shape[2])])

# Reshape y_test to 2D before applying LabelBinarizer
y_test_2d = y_test.reshape((y_test.shape[0], -1))

# Use LabelBinarizer for multiclass/multilabel problems
label_binarizer = LabelBinarizer()
y_test_encoded = label_binarizer.fit_transform(y_test_2d)

# Calculate AUC for each class
auc_per_class = roc_auc_score(y_test_encoded, y_pred, average=None)
print(f"AUC for each class: {auc_per_class}")

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
n_classes = y_test_encoded.shape[1]

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_encoded[:, i], y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curve for each class
plt.figure()
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], lw=2, label=f'Class {label_binarizer.classes_[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc='lower right')
plt.show()

# Decode predicted classes
y_pred_classes = np.argmax(y_pred, axis=-1)  # Use axis=-1 to ensure it works for the last axis
y_pred_decoded = label_encoder.inverse_transform(y_pred_classes)

# Decode true labels
y_test_decoded = label_encoder.inverse_transform(y_test_encoded.argmax(axis=1))

# Now, you can use y_test_decoded and y_pred_decoded for the confusion matrix
cm = confusion_matrix(y_test_decoded, y_pred_decoded)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_)
disp.plot(cmap=plt.cm.Reds, values_format="d")
plt.title("Confusion Matrix")
plt.show()

# Print test accuracy
test_acc = np.mean(y_pred_decoded == y_test_decoded)
print(f'Test accuracy: {test_acc}')
```

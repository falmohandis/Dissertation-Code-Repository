```python
def plot_clusters_3d(X, y_kmeans, cluster_centers, feature_names):
    num_features = X.shape[1]
    num_cols = 2
    num_rows = num_features // num_cols + (1 if num_features % num_cols != 0 else 0)
    fig = plt.figure(figsize=(12, 8*num_rows))
    for i in range(num_features):
        ax = fig.add_subplot(num_rows, num_cols, i+1, projection='3d')
        ax.scatter(X[:, i], X[:, (i+1)%num_features], X[:, (i+2)%num_features], c=y_kmeans, s=50, cmap='viridis')
        ax.scatter(cluster_centers[:, i], cluster_centers[:, (i+1)%num_features], cluster_centers[:, (i+2)%num_features], c='red', s=200, alpha=0.75)
        ax.set_xlabel(feature_names[i])
        ax.set_ylabel(feature_names[(i+1)%num_features])
        ax.set_zlabel(feature_names[(i+2)%num_features])
        ax.set_title(f'K-means Clustering\n({feature_names[i]} vs {feature_names[(i+1)%num_features]} vs {feature_names[(i+2)%num_features]})')
    plt.tight_layout()
    plt.show()

# def plot_clusters_2d(X, y_kmeans, cluster_centers, feature_names):
#     num_features = X.shape[1]
#     num_cols = 2
#     num_rows = num_features // num_cols + (1 if num_features % num_cols != 0 else 0)
#     fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12, 8*num_rows))
#     for i, ax in enumerate(axes.flat):
#         if i < num_features:
#             ax.scatter(X[:, i], X[:, (i+1)%num_features], c=y_kmeans, s=50, cmap='viridis')
#             ax.scatter(cluster_centers[:, i], cluster_centers[:, (i+1)%num_features], c='red', s=200, alpha=0.75)
#             ax.set_xlabel(feature_names[i])
#             ax.set_ylabel(feature_names[(i+1)%num_features])
#             ax.set_title(f'K-means Clustering\n({feature_names[i]} vs {feature_names[(i+1)%num_features]})')
#     plt.tight_layout()
#     plt.show()

def plot_clusters_2d(X, y_kmeans, cluster_centers, feature_names):
    num_features = X.shape[1]
    fig, axes = plt.subplots(nrows=num_features, ncols=num_features, figsize=(12, 12))
    for i in range(num_features):
        for j in range(num_features):
            if i != j:
                ax = axes[i, j]
                ax.scatter(X[:, i], X[:, j], c=y_kmeans, s=50, cmap='viridis')
                ax.scatter(cluster_centers[:, i], cluster_centers[:, j], c='red', s=200, alpha=0.75)
                ax.set_xlabel(feature_names[i])
                ax.set_ylabel(feature_names[j])
                ax.set_title(f'K-means Clustering\n({feature_names[i]} vs {feature_names[j]})')
    plt.tight_layout()
    plt.show()

```

```python

# Read data from Excel files in the current directory, excluding temporary or hidden files
file_names = [file for file in os.listdir('.') if file.endswith('.xlsx') and not file.startswith('~$')]

dfs = []
for file_name in file_names:
    xls = pd.ExcelFile(file_name)
    skip_file = False
    for sheet_name in xls.sheet_names:
        df = xls.parse(sheet_name)
        # Check for NaN values in the dataframe
        if df.isnull().values.any():
            print(f"NaN values found in '{file_name}'")
            skip_file = True
            break
        # Remove non-numeric characters from the data
        df = df.replace(regex=r'[^0-9.\-]', value='')
        dfs.append(df)
    if skip_file:
        continue

# Aggregate data from all dataframes into one dataframe
all_data = pd.concat(dfs, ignore_index=True)

# If custom feature names are specified, use only those columns
if default_feature_names is None:
    feature_names = all_data.columns.tolist()
else:
    feature_names = default_feature_names

# Extract features from the aggregated dataframe
X = all_data[feature_names].values.astype(float)  # Convert data to float after removing non-numeric characters

# Drop rows containing NaN values
X = X[~np.isnan(X).any(axis=1)]







# Apply K-means clustering
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
cluster_centers = kmeans.cluster_centers_

# Visualize the clusters in 3D
plot_clusters_3d(X, y_kmeans, cluster_centers, feature_names)

# Visualize the clusters in 2D
plot_clusters_2d(X, y_kmeans, cluster_centers, feature_names)
```

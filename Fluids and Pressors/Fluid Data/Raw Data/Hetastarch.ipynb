```python
import os
import pandas as pd
import numpy as np

def process_excel(filename, df):
    # Ensure required columns exist
    if 'Study Time(sec)' not in df.columns or 'Blood Bolus Number' not in df.columns:
        raise ValueError(f"Missing required column(s) in {filename}")

    # Convert and clean time column
    df['Study Time(sec)'] = pd.to_numeric(df['Study Time(sec)'], errors='coerce')
    df['Study Time(sec)'].fillna(method='ffill', inplace=True)
    df = df.sort_values('Study Time(sec)').reset_index(drop=True)
    df['Time_min'] = df['Study Time(sec)'] / 60.0

    # Get column as integer (handles floats safely)
    df['Blood Bolus Number'] = pd.to_numeric(df['Blood Bolus Number'], errors='coerce').fillna(0).astype(int)

    segments = []
    prev_val = df.loc[0, 'Blood Bolus Number']
    start_time = df.loc[0, 'Time_min']

    for i in range(1, len(df)):
        current_val = df.loc[i, 'Blood Bolus Number']

        # When the bolus number changes, close previous segment
        if current_val != prev_val:
            end_time = df.loc[i - 1, 'Time_min']
            duration = end_time - start_time
            segments.append({
                'File': filename,
                'Treatment Number': prev_val,
                'Start (min)': round(start_time, 3),
                'Stop (min)': round(end_time, 3),
                'Duration (min)': round(duration, 3)
            })
            # Start new segment
            start_time = df.loc[i, 'Time_min']
            prev_val = current_val

    # Handle the last segment
    end_time = df.loc[len(df) - 1, 'Time_min']
    duration = end_time - start_time
    segments.append({
        'File': filename,
        'Treatment Number': prev_val,
        'Start (min)': round(start_time, 3),
        'Stop (min)': round(end_time, 3),
        'Duration (min)': round(duration, 3)
    })

    return segments


# -------- Main Loop --------
if __name__ == "__main__":
    parent_folder_path = os.getcwd()
    print(f"Working directory: {parent_folder_path}")

    all_segments = []

    for file in os.listdir(parent_folder_path):
        if not (file.endswith('.xlsx') or file.endswith('.xls')):
            continue
        filepath = os.path.join(parent_folder_path, file)

        try:
            df = pd.read_excel(filepath)
            print(f"Processing {file} ...")
            segments = process_excel(file, df)
            all_segments.extend(segments)
        except Exception as e:
            print(f"Error processing {file}: {e}")

    # Export results
    if all_segments:
        summary_df = pd.DataFrame(all_segments)
        output_file = os.path.join(parent_folder_path, "Blood_Bolus_Treatment_Segments.xlsx")
        summary_df.to_excel(output_file, index=False)
        print(f"\n✅ Segments summary saved to: {output_file}")
    else:
        print("\n⚠️ No segments found or no valid files processed.")
```

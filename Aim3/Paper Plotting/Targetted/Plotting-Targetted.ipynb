```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ------------------------
# 🔹 PLOT CONFIG (UNIFIED STYLE SYSTEM)
# ------------------------
PLOT_CONFIG = {
    "figsize": (18, 5),
    "dpi": 300,

    "axis_tick_size": 10,
    "label_size": 14,
    "title_size": 16,
    "legend_size": 12,

    "x_tick_interval": 15,
    "x_min": 0,
    "x_max": 240,

    "line_width": 2,
    "marker_size": 5,
    "fill_alpha": 0.2,

    "ref_line_y": 1,

    "hspace": 0.35,
    "wspace": 0.10,
}

# ------------------------
# Data columns
# ------------------------
file_path = 'Fractional_Increase_0.xlsx'

columns_to_analyze = [
    'Avg',
    'Max',
    'Min',
    'Heart Rate',
    'Pseudo Cardiac Output'
]

hemorrhage_levels = [10, 20, 30]

color_palette = ['black', 'blue', 'red', 'purple', 'orange', 'cyan', 'magenta']

occlusion_label_map = {
    'No': 'No Aortic Occlusion',
    'No Occlusion': 'No Aortic Occlusion',
    'Partial': 'p-REBOA',
    'Partial Occlusion': 'p-REBOA',
    'Full': 'f-REBOA',
    'Full Occlusion': 'f-REBOA'
}

output_dir = 'Hemodynamics-Updated'
os.makedirs(output_dir, exist_ok=True)

# ------------------------
# Load data
# ------------------------
df = pd.read_excel(file_path)

order_of_time_points = sorted(df['Time Point'].dropna().unique())
order_of_time_points = [int(tp) for tp in order_of_time_points]

df['Time Point'] = df['Time Point'].apply(
    lambda x: x - 1 if x in order_of_time_points and x != 0 else x
)

df['Test Group'] = df['Occlusion Group']
df = df.sort_values(by=['Probe Location', 'Hemorrhage Level'])

probe_locations = sorted(df['Probe Location'].dropna().unique())

# ------------------------
# Plotting function
# ------------------------
def create_metric_multiplot(metric):

    n_rows = len(probe_locations)
    n_cols = len(hemorrhage_levels)

    # ✅ CHANGED: taller figure scaling
    fig = plt.figure(figsize=(
        PLOT_CONFIG["figsize"][0] * (n_cols / 3),
        PLOT_CONFIG["figsize"][1] * n_rows
    ), dpi=PLOT_CONFIG["dpi"])

    axes = np.empty((n_rows, n_cols), dtype=object)

    for i in range(n_rows):
        for j in range(n_cols):
            axes[i, j] = fig.add_subplot(n_rows, n_cols, i * n_cols + j + 1)

    handles, labels = [], []

    for row_idx, probe_location in enumerate(probe_locations):
        for col_idx, hemorrhage_level in enumerate(hemorrhage_levels):

            ax = axes[row_idx, col_idx]

            subset_df = df[
                (df['Probe Location'] == probe_location) &
                (df['Hemorrhage Level'] == hemorrhage_level)
            ]

            # ✅ CHANGED: title appears on every graph, even if empty
            ax.set_title(
                f"{int(hemorrhage_level)}% Hemorrhage",
                fontsize=PLOT_CONFIG["label_size"],
                pad=18
            )

            if subset_df.empty:
                ax.text(
                    0.5, 0.5,
                    "No Data",
                    ha="center",
                    va="center",
                    transform=ax.transAxes,
                    fontsize=PLOT_CONFIG["label_size"]
                )

                ax.set_xticks(np.arange(
                    PLOT_CONFIG["x_min"],
                    PLOT_CONFIG["x_max"] + 1,
                    PLOT_CONFIG["x_tick_interval"]
                ))
                ax.set_xlim(PLOT_CONFIG["x_min"], PLOT_CONFIG["x_max"])

                # ✅ CHANGED: force x-axis labels on empty plots too
                ax.tick_params(
                    axis='x',
                    labelbottom=True,
                    labeltop=False,
                    labelsize=PLOT_CONFIG["axis_tick_size"]
                )
                ax.tick_params(
                    axis='y',
                    labelsize=PLOT_CONFIG["axis_tick_size"]
                )

                ax.set_xlabel("Time (mins)", fontsize=PLOT_CONFIG["label_size"])
                ax.grid(True, linestyle='--', linewidth=0.7)
                continue

            numeric_columns = subset_df.select_dtypes(include=['number']).columns

            mean_df = subset_df.groupby(
                ['Time Point', 'Test Group'],
                as_index=False
            )[numeric_columns].mean()

            std_df = subset_df.groupby(
                ['Time Point', 'Test Group'],
                as_index=False
            )[numeric_columns].std()

            for i, test_group in enumerate(subset_df['Test Group'].unique()):
                mean_subset = mean_df[mean_df['Test Group'] == test_group]
                std_subset = std_df[std_df['Test Group'] == test_group]

                if mean_subset.empty:
                    continue

                color = color_palette[i % len(color_palette)]
                legend_label = occlusion_label_map.get(test_group, test_group)

                ax.plot(
                    mean_subset['Time Point'],
                    mean_subset[metric],
                    linewidth=PLOT_CONFIG["line_width"],
                    marker='o',
                    markersize=PLOT_CONFIG["marker_size"],
                    color=color,
                    label=legend_label
                )

                ax.fill_between(
                    mean_subset['Time Point'],
                    mean_subset[metric] - std_subset[metric],
                    mean_subset[metric] + std_subset[metric],
                    alpha=PLOT_CONFIG["fill_alpha"],
                    color=color
                )

                if legend_label not in labels:
                    handles.append(ax.lines[-1])
                    labels.append(legend_label)

            # reference line
            ax.axhline(
                y=PLOT_CONFIG["ref_line_y"],
                linestyle='--',
                linewidth=1
            )

            if col_idx == 0:
                ax.set_ylabel(
                    str(probe_location).replace("_", " ").title(),
                    fontsize=PLOT_CONFIG["label_size"]
                )

            # ✅ CHANGED: every graph gets an x-axis label
            ax.set_xlabel("Time (mins)", fontsize=PLOT_CONFIG["label_size"])

            # ticks
            ax.set_xticks(np.arange(
                PLOT_CONFIG["x_min"],
                PLOT_CONFIG["x_max"] + 1,
                PLOT_CONFIG["x_tick_interval"]
            ))

            ax.set_xlim(PLOT_CONFIG["x_min"], PLOT_CONFIG["x_max"])

            # ✅ CHANGED: force x-axis tick labels to show on every subplot
            ax.tick_params(
                axis='x',
                labelbottom=True,
                labeltop=False,
                bottom=True,
                top=False,
                labelsize=PLOT_CONFIG["axis_tick_size"]
            )

            ax.tick_params(
                axis='y',
                labelsize=PLOT_CONFIG["axis_tick_size"]
            )

            ax.grid(True, linestyle='--', linewidth=0.7)

    # title
    fig.suptitle(metric, fontsize=PLOT_CONFIG["label_size"] + 4)

    # legend optional
    # fig.legend(
    #     handles,
    #     labels,
    #     loc='center right',
    #     fontsize=PLOT_CONFIG["legend_size"]
    # )

    # ✅ CHANGED: more spacing between graphs
    plt.subplots_adjust(
        hspace=PLOT_CONFIG["hspace"],
        wspace=PLOT_CONFIG["wspace"],
        top=0.94,
        bottom=0.05,
        left=0.05,
        right=0.98
    )

    filename = os.path.join(output_dir, f"{metric}_multiplot.png")
    plt.savefig(filename, dpi=PLOT_CONFIG["dpi"])
    plt.close()


# ------------------------
# Run
# ------------------------
for metric in columns_to_analyze:
    create_metric_multiplot(metric)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re

# ------------------------
# 🔹 PLOT CONFIG (UNIFIED STYLE SYSTEM)
# ------------------------
PLOT_CONFIG = {
    "figsize": (18, 5),
    "dpi": 300,

    "axis_tick_size": 10,
    "label_size": 14,
    "title_size": 16,
    "legend_size": 12,

    "x_tick_interval": 15,
    "x_min": 0,
    "x_max": 240,

    "line_width": 2,
    "marker_size": 5,
    "fill_alpha": 0.2,

    "ref_line_y": 1,

    "hspace": 0.35,
    "wspace": 0.10,
}

# ------------------------
# Terms to plot (ROWS)
# ------------------------
terms_to_plot = [
    "Cardiac Output",
    "ESP"
]

# ------------------------
# Style definitions
# ------------------------
marker_info = {
    "None":    {"marker": "o", "color": "black"},
    "Partial": {"marker": "o", "color": "blue"},
    "Full":    {"marker": "o", "color": "red"}
}

legend_order = ["None", "Partial", "Full"]

occlusion_label_map = {
    "None": "No Aortic Occlusion",
    "No": "No Aortic Occlusion",
    "Partial": "p-REBOA",
    "Full": "f-REBOA"
}

# ------------------------
# Load and preprocess data
# ------------------------
file_path = 'OG-PV_Fractional_Increase_T0-1.xlsx'
df = pd.read_excel(file_path)

df = df.drop(columns=["Weight (kg)", "Gender"])

# 🔹 REQUIRED: define your time_point_order
time_point_order = df['Time Point'].astype(str).unique().tolist()

df['Time Point Raw'] = df['Time Point'].astype(str)
df = df[df['Time Point Raw'].isin(time_point_order)]

# ------------------------
# Time extraction
# ------------------------
def extract_time_point_number(time_point):
    match = re.match(r"T(\d+)-(\d+)", str(time_point))
    if match:
        first = int(match.group(1))
        second = int(match.group(2))

        if time_point in ["T29-30", "T239-240"]:
            return second

        return first

    return None

df['Time Point'] = df['Time Point Raw'].apply(extract_time_point_number)

# ------------------------
# Treatment parsing
# ------------------------
def extract_treatment_details(treatment):
    match = re.match(r"(\d+)%\s*(\w+)", str(treatment))
    if match:
        return match.group(1), match.group(2)

    return None, None

df[['Treatment Percentage', 'Treatment Group']] = df['Treatment Group'].apply(
    lambda x: pd.Series(extract_treatment_details(x))
)

df = df.sort_values('Time Point')

# 🔹 Hemorrhage levels (COLUMNS)
hemorrhage_levels = sorted(df['Hemorrhage Level'].dropna().unique())

# ------------------------
# Output folder
# ------------------------
output_folder = "Pressure-Volume-Updated"
os.makedirs(output_folder, exist_ok=True)

# ------------------------
# Plotting function
# ------------------------
def create_multiterm_panel():

    numeric_columns = df.select_dtypes(include=['number']).columns
    valid_terms = [t for t in terms_to_plot if t in numeric_columns]

    n_rows = len(valid_terms)
    n_cols = len(hemorrhage_levels)

    # ✅ CHANGED: manual figure creation for better control
    fig = plt.figure(figsize=(
        PLOT_CONFIG["figsize"][0] * (n_cols / 3),
        PLOT_CONFIG["figsize"][1] * n_rows
    ), dpi=PLOT_CONFIG["dpi"])

    axes = np.empty((n_rows, n_cols), dtype=object)

    for i in range(n_rows):
        for j in range(n_cols):
            axes[i, j] = fig.add_subplot(n_rows, n_cols, i * n_cols + j + 1)

    handles, labels = [], []

    for row_idx, term in enumerate(valid_terms):
        for col_idx, hemorrhage_level in enumerate(hemorrhage_levels):

            ax = axes[row_idx, col_idx]

            # ✅ CHANGED: every subplot gets a hemorrhage title
            ax.set_title(
                f"{int(hemorrhage_level)}% Hemorrhage",
                fontsize=PLOT_CONFIG["label_size"],
                pad=18
            )

            subset_df = df[df['Hemorrhage Level'] == hemorrhage_level]

            if subset_df.empty:
                ax.text(
                    0.5, 0.5,
                    "No Data",
                    ha="center",
                    va="center",
                    transform=ax.transAxes,
                    fontsize=PLOT_CONFIG["label_size"]
                )

                ax.set_xticks(np.arange(
                    PLOT_CONFIG["x_min"],
                    PLOT_CONFIG["x_max"] + 1,
                    PLOT_CONFIG["x_tick_interval"]
                ))

                ax.set_xlim(PLOT_CONFIG["x_min"], PLOT_CONFIG["x_max"])

                # ✅ CHANGED: force x-axis on empty graphs too
                ax.set_xlabel("Time (mins)", fontsize=PLOT_CONFIG["label_size"])
                ax.tick_params(
                    axis='x',
                    labelbottom=True,
                    labeltop=False,
                    bottom=True,
                    top=False,
                    labelsize=PLOT_CONFIG["axis_tick_size"]
                )
                ax.tick_params(
                    axis='y',
                    labelsize=PLOT_CONFIG["axis_tick_size"]
                )

                ax.grid(True, linestyle='--', linewidth=0.7)
                continue

            numeric_columns = subset_df.select_dtypes(include=['number']).columns

            mean_df = subset_df.groupby(
                ['Treatment Group', 'Time Point'],
                as_index=False
            )[numeric_columns].mean()

            std_df = subset_df.groupby(
                ['Treatment Group', 'Time Point'],
                as_index=False
            )[numeric_columns].std()

            mean_df['Time Point'] = mean_df['Time Point'].astype(int)
            std_df['Time Point'] = std_df['Time Point'].astype(int)

            for treatment_key in legend_order:
                mean_subset = mean_df[mean_df['Treatment Group'] == treatment_key]
                std_subset = std_df[std_df['Treatment Group'] == treatment_key]

                if mean_subset.empty:
                    continue

                color = marker_info[treatment_key]["color"]
                marker = marker_info[treatment_key]["marker"]

                legend_label = occlusion_label_map.get(treatment_key, treatment_key)

                line, = ax.plot(
                    mean_subset['Time Point'],
                    mean_subset[term],
                    marker=marker,
                    markersize=PLOT_CONFIG["marker_size"],
                    linewidth=PLOT_CONFIG["line_width"],
                    color=color,
                    label=legend_label
                )

                ax.fill_between(
                    mean_subset['Time Point'],
                    mean_subset[term] - std_subset[term],
                    mean_subset[term] + std_subset[term],
                    alpha=PLOT_CONFIG["fill_alpha"],
                    color=color
                )

                if legend_label not in labels:
                    handles.append(line)
                    labels.append(legend_label)

            # Reference line
            ax.axhline(
                y=PLOT_CONFIG["ref_line_y"],
                linestyle='--',
                linewidth=1
            )

            # Row labels
            if col_idx == 0:
                formatted_term = str(term).replace("_", " ").title()
                ax.set_ylabel(
                    formatted_term,
                    fontsize=PLOT_CONFIG["label_size"]
                )

            # ✅ CHANGED: every graph gets x-axis label
            ax.set_xlabel("Time (mins)", fontsize=PLOT_CONFIG["label_size"])

            # X ticks
            if PLOT_CONFIG["x_min"] is not None and PLOT_CONFIG["x_max"] is not None:
                start_tick = PLOT_CONFIG["x_min"]
                end_tick = PLOT_CONFIG["x_max"]
            else:
                min_time = mean_df['Time Point'].min()
                max_time = mean_df['Time Point'].max()
                start_tick = (
                    min_time // PLOT_CONFIG["x_tick_interval"]
                ) * PLOT_CONFIG["x_tick_interval"]
                end_tick = (
                    (max_time // PLOT_CONFIG["x_tick_interval"]) + 1
                ) * PLOT_CONFIG["x_tick_interval"]

            ax.set_xticks(np.arange(
                start_tick,
                end_tick + 1,
                PLOT_CONFIG["x_tick_interval"]
            ))

            ax.set_xlim(start_tick, end_tick)

            # ✅ CHANGED: force x-axis tick labels to appear on every subplot
            ax.tick_params(
                axis='x',
                labelbottom=True,
                labeltop=False,
                bottom=True,
                top=False,
                labelsize=PLOT_CONFIG["axis_tick_size"]
            )

            ax.tick_params(
                axis='y',
                labelsize=PLOT_CONFIG["axis_tick_size"]
            )

            ax.grid(True, linestyle='--', linewidth=0.7)

    # Shared legend optional
    # fig.legend(
    #     handles,
    #     labels,
    #     loc='center right',
    #     fontsize=PLOT_CONFIG["legend_size"]
    # )

    # ✅ CHANGED: more spacing between plots
    plt.subplots_adjust(
        hspace=PLOT_CONFIG["hspace"],
        wspace=PLOT_CONFIG["wspace"],
        top=0.94,
        bottom=0.05,
        left=0.06,
        right=0.98
    )

    filename = os.path.join(output_folder, "Multiterm_Panel.png")
    plt.savefig(filename, dpi=PLOT_CONFIG["dpi"])
    plt.close()

# ------------------------
# Run
# ------------------------
create_multiterm_panel()
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# 🔹 GLOBAL PLOT CONFIG (MATCHED STYLE)
# ============================================================

TITLE_FONTSIZE = 20
LABEL_FONTSIZE = 16
TICK_FONTSIZE = 14
LEGEND_FONTSIZE = 14

output_file = "Hemorrhage_Occlusion_3x3_grid.png"

y_axis_labels = {
    "Plasmalyte": "Plasmalyte (mL)",
    "Cumulative Vasopressin": "Cumulative Vasopressin (pg/mL)",
    "Norepi": "Norepinephrine (ng/kg/min)"
}

# ============================================================
# Data Loading
# ============================================================

df = pd.read_excel(
    'Experiment Tracker.xlsx',
    sheet_name=['T30 categorization', 'T60 categorization']
)

group_10 = set(df['T30 categorization']['Group_10'].dropna())
group_20 = set(df['T30 categorization']['Group_20'].dropna())
group_30 = set(df['T30 categorization']['Group_30'].dropna())

full_occlusion = set(df['T60 categorization']['Full_Occlusion'].dropna())
partial_occlusion = set(df['T60 categorization']['Partial_Occlusion'].dropna())
no_occlusion = set(df['T60 categorization']['No_Occlusion'].dropna())

group_combinations = {
    "10": {
        "No": group_10.intersection(no_occlusion),
        "Partial": group_10.intersection(partial_occlusion),
        "Full": group_10.intersection(full_occlusion)
    },
    "20": {
        "No": group_20.intersection(no_occlusion),
        "Partial": group_20.intersection(partial_occlusion),
        "Full": group_20.intersection(full_occlusion)
    },
    "30": {
        "No": group_30.intersection(no_occlusion),
        "Partial": group_30.intersection(partial_occlusion),
        "Full": group_30.intersection(full_occlusion)
    }
}

xls = pd.ExcelFile("Processed_Data_with_Vasopressin.xlsx")
erne_sheets = [s for s in xls.sheet_names if s.startswith("ERNE")]

data_dict = {}
for sheet in erne_sheets:
    df_sheet = xls.parse(sheet)
    new_sheet_name = sheet.replace("-", "")

    required_columns = [
        "Time", "Cumulative Vasopressin", "Plasmalyte", "Norepi",
        "Prox_Sys_PowerLab_PC", "Prox_Dia_PowerLab_PC", "Prox_Mean_PowerLab_PC",
        "Dist_Sys_PowerLab_PC", "Dist_Dia_PowerLab_PC", "Dist_Mean_PowerLab_PC",
        "Dist_Sys_LPF_PowerLab_PC", "Dist_Dia_LPF_PowerLab_PC", "Dist_Mean_LPF_PowerLab_PC"
    ]

    df_sheet = df_sheet[required_columns]
    df_sheet = df_sheet[df_sheet["Time"] <= 240]

    data_dict[new_sheet_name] = df_sheet

# ============================================================
# Helper
# ============================================================

def calculate_mean_std(subjects, value_col, time_points):
    group_values = {}

    for sub in subjects:
        if sub in data_dict:
            df_local = data_dict[sub]
            df_filt = df_local[df_local["Time"].isin(time_points)]

            for _, row in df_filt.iterrows():
                t = row["Time"]
                group_values.setdefault(t, []).append(row[value_col])

    if not group_values:
        return [], [], []

    times = sorted(group_values.keys())
    means = [np.mean(group_values[t]) for t in times]
    stds = [np.std(group_values[t]) for t in times]

    return times, means, stds

# ============================================================
# Plotting
# ============================================================

time_points = [0, 5, 10, 15, 20, 25, 30, 60, 65, 75, 85, 120, 180, 240]

columns_to_graph = ["Plasmalyte", "Cumulative Vasopressin", "Norepi"]
hemorrhage_order = ["10", "20", "30"]
occlusion_order = ["No", "Partial", "Full"]

occlusion_styles = {
    "No": {"color": "black", "marker": "o"},
    "Partial": {"color": "blue", "marker": "o"},
    "Full": {"color": "red", "marker": "o"}
}

occlusion_label_map = {
    "No": "No REBOA",
    "Partial": "p-REBOA",
    "Full": "f-REBOA"
}

# ============================================================
# Figure sizing (MATCHED SYSTEM STYLE)
# ============================================================

n_rows = len(columns_to_graph)
n_cols = len(hemorrhage_order)

fig, axes = plt.subplots(
    n_rows,
    n_cols,
    figsize=(6 * n_cols, 3.8 * n_rows),
    sharex=False,
    sharey=False
)

# ============================================================
# Plotting loop
# ============================================================

for row_idx, column in enumerate(columns_to_graph):
    for col_idx, hemorrhage_level in enumerate(hemorrhage_order):

        ax = axes[row_idx, col_idx]

        for occlusion_type in occlusion_order:
            subjects = group_combinations[hemorrhage_level].get(occlusion_type, set())

            if not subjects:
                continue

            times, means, stds = calculate_mean_std(subjects, column, time_points)

            if times:
                style = occlusion_styles[occlusion_type]

                ax.plot(
                    times,
                    means,
                    label=occlusion_label_map[occlusion_type],
                    color=style["color"],
                    marker=style["marker"],
                    linewidth=2
                )

                ax.fill_between(
                    times,
                    np.array(means) - np.array(stds),
                    np.array(means) + np.array(stds),
                    color=style["color"],
                    alpha=0.2
                )

        # ------------------------
        # Labels / formatting
        # ------------------------

        ax.set_xlabel("Time (min)", fontsize=LABEL_FONTSIZE)
        ax.set_ylabel(y_axis_labels[column], fontsize=LABEL_FONTSIZE)

        ax.tick_params(axis='both', labelsize=TICK_FONTSIZE)
        ax.set_xticks(range(0, 241, 30))

        if row_idx == 0:
            ax.set_title(f"{hemorrhage_level}% Hemorrhage", fontsize=TITLE_FONTSIZE)

        ax.grid(True, linestyle="--", linewidth=0.5)

# ============================================================
# Legend (global)
# ============================================================

handles = [
    plt.Line2D(
        [0], [0],
        color=occlusion_styles[o]["color"],
        marker=occlusion_styles[o]["marker"],
        label=occlusion_label_map[o],
        linewidth=2
    )
    for o in occlusion_order
]

fig.legend(
    handles=handles,
    loc='upper right',
    bbox_to_anchor=(0.16, 0.90),
    fontsize=LEGEND_FONTSIZE,
    title="Occlusion Type",
    title_fontsize=LABEL_FONTSIZE
)

# ============================================================
# FINAL LAYOUT (MATCHED ACROSS ALL YOUR FIGURES)
# ============================================================

plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.savefig(output_file, dpi=300)
plt.close()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------
# Parameters
# ------------------------
file_path = 'Fractional_Increase_0.xlsx'

AXIS_TICK_SIZE = 16
TITLE_SIZE = 20
LABEL_SIZE = 18

color_palette = ['#E69F00', '#009E73', '#CC79A7']  # Orange, Teal, Purple


marker_map = {
    10: 'o',   # circle
    20: 'o',   # square
    30: 'o'    # triangle
}

output_dir = 'Hemorrhage Comparison'
os.makedirs(output_dir, exist_ok=True)

# ------------------------
# Load and preprocess data
# ------------------------
df = pd.read_excel(file_path)



df['Time Point'] = df['Time Point'].astype(int)
df = df[df['Time Point'].astype(str).isin(order_of_time_points)]

df['Test Group'] = df['Occlusion Group']
df = df.sort_values(by=['Hemorrhage Level', 'Gender'])

# ------------------------
# Plotting function
# ------------------------
def create_combined_plot(probe_location):
    subset_df = df[df['Probe Location'] == probe_location]
    if subset_df.empty:
        print(f'No data available for Probe Location: {probe_location}')
        return

    hemorrhage_levels = [10, 20, 30]

    for metric in ['Avg', 'Max', 'Min']:
        plt.figure(figsize=(10, 6))

        for i, hemorrhage_level in enumerate(hemorrhage_levels):
            level_df = subset_df[subset_df['Hemorrhage Level'] == hemorrhage_level]
            if level_df.empty:
                continue

            mean_df = level_df.groupby('Time Point')[metric].mean()
            std_df = level_df.groupby('Time Point')[metric].std()

            time_points = [int(tp) for tp in order_of_time_points if int(tp) in mean_df.index]
            mean_values = [mean_df[tp] for tp in time_points]
            std_values = [std_df[tp] for tp in time_points]

            plt.plot(
                time_points,
                mean_values,
                marker=marker_map[hemorrhage_level],
                markersize=8,
                linewidth=2,
                color=color_palette[i],
                label=f'{hemorrhage_level}% Hemorrhage'
            )

            plt.fill_between(
                time_points,
                [m - s for m, s in zip(mean_values, std_values)],
                [m + s for m, s in zip(mean_values, std_values)],
                color=color_palette[i],
                alpha=0.2
            )

        plt.axhline(y=1, color="gray", linestyle="--", linewidth=1)
        plt.xlabel("Time (mins)", fontsize=LABEL_SIZE)
        plt.ylabel(f"Normalized {probe_location} - {metric}", fontsize=LABEL_SIZE)
        plt.title(f"{probe_location} - {metric}", fontsize=TITLE_SIZE)
        plt.grid(True, linestyle='--', linewidth=0.5)
        plt.xticks(fontsize=AXIS_TICK_SIZE)
        plt.yticks(fontsize=AXIS_TICK_SIZE)
        plt.legend(fontsize=12)

        plt.tight_layout()
        plt.savefig(
            os.path.join(output_dir, f'Probe_{probe_location}_{metric}_combined.png'),
            dpi=300
        )
        plt.close()

# ------------------------
# Run for all probe locations
# ------------------------
for probe_location in df['Probe Location'].unique():
    create_combined_plot(probe_location)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re

AXIS_TICK_SIZE = 24
TITLE_SIZE = 24
LABEL_SIZE = 24
LEGEND_SIZE = 22

# Control x-axis tick interval here
X_TICK_INTERVAL = 1   # spacing in index units (keep as is since boxes are equidistant in these plots)

marker_info = {
    "None":    {"color": "black"},
    "Partial": {"color": "blue"},
    "Full":    {"color": "red"}
}


# marker_info = {
#     "None":    {"color":"#E69F00"},   # Orange
#     "Partial": {"color": "#009E73"},   # Teal/Green
#     "Full":    {"color": "#CC79A7"}    # Purple/Magenta
# }


legend_order = ["None", "Partial", "Full"]

# Forced legend naming
occlusion_label_map = {
    "None": "No Aortic Occlusion",
    "No": "No Aortic Occlusion",
    "Partial": "p-REBOA",
    "Full": "f-REBOA"
}

# ------------------------
# User-specified time points
# ------------------------
time_point_order = [
    "T30-T31",
    "T60-61",
    "T75-76",
    "T85-86",
    "T120-121",
    "T180-181",
    "T239-240"
]

time_value_map = {
    "T30-T31": 30,
    "T60-61": 60,
    "T75-76": 75,
    "T85-86": 85,
    "T120-121": 120,
    "T180-181": 180,
    "T239-240": 240
}

time_index_map = {tp: i for i, tp in enumerate(time_point_order)}

# ------------------------
# Load and preprocess data
# ------------------------
file_path = 'PV_Fractional_Increase_T0-1.xlsx'
df = pd.read_excel(file_path)
df = df.drop(columns=["Weight (kg)", "Gender"])
df['Time Point Raw'] = df['Time Point'].astype(str)

df = df[df['Time Point Raw'].isin(time_point_order)]
df['Time Index'] = df['Time Point Raw'].map(time_index_map)

def extract_treatment_details(treatment):
    match = re.match(r"(\d+)%\s*(\w+)", str(treatment))
    if match:
        return match.group(1), match.group(2)
    return None, None

df[['Treatment Percentage', 'Treatment Group']] = df['Treatment Group'].apply(
    lambda x: pd.Series(extract_treatment_details(x))
)

df = df.sort_values('Time Index')

# ------------------------
# Create output folder
# ------------------------
output_folder = "PV Box Plots"
os.makedirs(output_folder, exist_ok=True)

# ------------------------
# Plotting function
# ------------------------
def create_panel_plots():
    numeric_columns = df.select_dtypes(include=['number']).columns
    hemorrhage_levels = sorted(df['Hemorrhage Level'].unique())

    for term in [t for t in terms_to_plot if t in numeric_columns]:
        fig, axes = plt.subplots(
            1,
            len(hemorrhage_levels),
            figsize=(36, 6),
            sharey=True
        )

        if len(hemorrhage_levels) == 1:
            axes = [axes]

        handles, labels = [], []

        for ax, hemorrhage_level in zip(axes, hemorrhage_levels):
            subset_df = df[df['Hemorrhage Level'] == hemorrhage_level]

            if subset_df.empty:
                ax.set_title(f"{hemorrhage_level}% Hemorrhage - No Data", fontsize=TITLE_SIZE)
                continue

            n_treatments = len(legend_order)
            box_width = 0.2
            offsets = [(i - (n_treatments-1)/2) * box_width*1.5 for i in range(n_treatments)]

            for i, treatment_key in enumerate(legend_order):
                treatment_df = subset_df[subset_df['Treatment Group'] == treatment_key]
                if treatment_df.empty:
                    continue

                color = marker_info[treatment_key]["color"]
                legend_label = occlusion_label_map.get(treatment_key, treatment_key)

                data, positions = [], []

                for tp in time_point_order:
                    vals = treatment_df[treatment_df['Time Point Raw'] == tp][term].dropna()
                    if not vals.empty:
                        data.append(vals)
                        positions.append(time_index_map[tp] + offsets[i])

                if not data:
                    continue

                bp = ax.boxplot(
                    data,
                    positions=positions,
                    widths=box_width,
                    patch_artist=True,
                    showfliers=False
                )

                for box in bp['boxes']:
                    box.set_facecolor(color)
                    box.set_alpha(0.6)
                    box.set_linewidth(2)

                for element in ['whiskers', 'caps', 'medians']:
                    for item in bp[element]:
                        item.set_linewidth(2)

                if legend_label not in labels:
                    handles.append(bp['boxes'][0])
                    labels.append(legend_label)

            ax.axhline(y=1, color='gray', linestyle='--', linewidth=1)

            # 🔹 Title cleaned
            ax.set_title(f"{hemorrhage_level}% Hemorrhage", fontsize=TITLE_SIZE)

            # 🔹 Custom tick interval
            xtick_positions = np.arange(
                0,
                len(time_point_order),
                X_TICK_INTERVAL
            )

            ax.set_xticks(xtick_positions)
            ax.set_xticklabels(
                [time_value_map[time_point_order[int(i)]] for i in xtick_positions]
            )

            ax.set_xlabel("Time (mins)", fontsize=LABEL_SIZE)

            # ax.grid(True, axis='y', linestyle='--', linewidth=0.5)
            ax.tick_params(axis='both', labelsize=AXIS_TICK_SIZE)

        # 🔹 Y-axis forced
        # axes[0].set_ylabel("Normalized Renal Flow", fontsize=LABEL_SIZE)
        # Automatically clean and format term name
        formatted_term = term.replace("_", " ")

        axes[0].set_ylabel(f"Normalized {formatted_term}", fontsize=LABEL_SIZE)


        # 🔹 Legend standardized
        axes[-1].legend(
            handles,
            labels,
            title="Intervention Type",
            loc='center left',
            bbox_to_anchor=(1.02, 0.5),
            fontsize=LEGEND_SIZE,
            title_fontsize=LEGEND_SIZE,
            borderaxespad=0
        )

        plt.tight_layout()
        plt.savefig(
            os.path.join(output_folder, f"{term}_Panel.png"),
            dpi=300,
            bbox_inches='tight'
        )
        plt.close()

# ------------------------
# Run plotting
# ------------------------
create_panel_plots()
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# Parameters
# ============================================================

file_path = 'ERNE_TEG_cleaned.xlsx'

AXIS_TICK_SIZE = 16
TITLE_SIZE = 20
LABEL_SIZE = 18
PANEL_LABEL_SIZE = 18

output_dir = 'TEG'
os.makedirs(output_dir, exist_ok=True)

# ============================================================
# Output files
# ============================================================

output_image_1 = os.path.join(output_dir, "TEG_Image_1_R_Angle_G_3x3.png")
output_image_2 = os.path.join(output_dir, "TEG_Image_2_K_MA_LY30_3x3.png")

# ============================================================
# TEG metric layout
# ============================================================

image_1_metrics = [
    ("A", "R_time", "Reaction Time"),
    ("C", "Angle", "Alpha Angle"),
    ("E", "G", "Shear Modulus")
]

image_2_metrics = [
    ("B", "K_time", "Kinetic Time"),
    ("D", "MA", "Maximum Amplitude"),
    ("F", "LY30", "% of Clot Lysis 30 min post maximum amplitude")
]

# ============================================================
# Y-axis labels
# Parameter names are kept on the y-axis, not in the titles
# ============================================================

y_axis_labels = {
    "R_time": "Reaction Time",
    "K_time": "Kinetic Time",
    "Angle": "Alpha Angle",
    "MA": "Maximum Amplitude",
    "G": "Shear Modulus",
    "LY30": "% of Clot Lysis 30 min post maximum amplitude"
}

# ============================================================
# Y-axis limits
# Set min/max here for each graph.
#
# Use None if you want matplotlib to choose automatically.
# ============================================================

y_axis_limits = {
    "R_time": (None, None),
    "K_time": (None, None),

    # Based on feedback
    "Angle": (20, 85),
    "MA": (50, 85),
    "G": (5000, 18000),

    "LY30": (None, None)
}

# ============================================================
# Optional dotted vertical phase lines
#
# Edit these values based on your experimental phases.
# These will draw vertical dotted lines on every subplot.
# Set phase_lines = [] if you do not want them.
# ============================================================

phase_lines = [30, 60, 120, 180]

PHASE_LINE_COLOR = "gray"
PHASE_LINE_STYLE = ":"
PHASE_LINE_WIDTH = 1.5

# ============================================================
# Marker symbols and colors
# ============================================================

marker_info = {
    "No": {"marker": "o", "color": "black"},
    "Partial": {"marker": "o", "color": "blue"},
    "Full": {"marker": "o", "color": "red"}
}

legend_order = ["No", "Partial", "Full"]

# ============================================================
# Load and preprocess data
# ============================================================

df = pd.read_excel(file_path)

time_col = 'Time Point'
group_col = 'Occlusion Group'
level_col = 'Hemorrhage Level'

# Ensure numeric time
df[time_col] = pd.to_numeric(df[time_col], errors='coerce')
df = df.dropna(subset=[time_col])
df[time_col] = df[time_col].astype(float)

# Ensure numeric hemorrhage level
df[level_col] = pd.to_numeric(df[level_col], errors='coerce')

# Correct typo
df[group_col] = df[group_col].replace({
    "None Occlusion": "No Occlusion"
})

# Map occlusion group names to simplified plotting keys
def map_group_key(name):
    name = str(name)

    if "No" in name:
        return "No"
    elif "Partial" in name:
        return "Partial"
    elif "Full" in name:
        return "Full"

    return "No"

df["Group Key"] = df[group_col].apply(map_group_key)

# Make hemorrhage order explicit
hemorrhage_order = [10, 20, 30]

# ============================================================
# Plotting function
# ============================================================

def plot_teg_3x3(metric_layout, output_file):
    n_rows = len(metric_layout)
    n_cols = len(hemorrhage_order)

    fig, axes = plt.subplots(
        n_rows,
        n_cols,
        figsize=(18, 16),
        sharex=False,
        sharey=False
    )

    # Fixed margins keep both saved images and graph widths identical.
    # Increased top and hspace so labels can sit above graphs.
    fig.subplots_adjust(
        left=0.10,
        right=0.98,
        top=0.88,
        bottom=0.08,
        wspace=0.28,
        hspace=0.62
    )

    for row_idx, (panel_letter, metric, metric_name) in enumerate(metric_layout):
        for col_idx, hemorrhage_level in enumerate(hemorrhage_order):

            ax = axes[row_idx, col_idx]

            level_df = df[df[level_col] == hemorrhage_level]

            grouped_df = (
                level_df
                .groupby(["Group Key", time_col])[metric]
                .agg(["mean", "std"])
                .reset_index()
            )

            for key in legend_order:
                subset = grouped_df[grouped_df["Group Key"] == key]

                if subset.empty:
                    continue

                ax.plot(
                    subset[time_col],
                    subset["mean"],
                    marker=marker_info[key]["marker"],
                    markersize=8,
                    linewidth=2,
                    color=marker_info[key]["color"]
                )

                ax.fill_between(
                    subset[time_col],
                    subset["mean"] - subset["std"],
                    subset["mean"] + subset["std"],
                    alpha=0.2,
                    color=marker_info[key]["color"]
                )

            # ------------------------------------------------
            # Dotted vertical phase demarcation lines
            # ------------------------------------------------

            for phase_time in phase_lines:
                ax.axvline(
                    x=phase_time,
                    color=PHASE_LINE_COLOR,
                    linestyle=PHASE_LINE_STYLE,
                    linewidth=PHASE_LINE_WIDTH
                )

            # ------------------------------------------------
            # Column titles only on top row, pushed higher
            # ------------------------------------------------

            if row_idx == 0:
                ax.set_title(
                    f"{hemorrhage_level}% Hemorrhage",
                    fontsize=TITLE_SIZE,
                    pad=42
                )

            # ------------------------------------------------
            # Panel labels above each graph
            # ------------------------------------------------

            ax.text(
                0.00,
                1.10,
                f"{panel_letter}. {metric_name}",
                transform=ax.transAxes,
                fontsize=PANEL_LABEL_SIZE,
                fontweight="bold",
                va="bottom",
                ha="left",
                clip_on=False
            )

            # ------------------------------------------------
            # Axis labels
            # ------------------------------------------------

            ax.set_xlabel("Time", fontsize=LABEL_SIZE)

            if col_idx == 0:
                ax.set_ylabel(
                    y_axis_labels[metric],
                    fontsize=LABEL_SIZE
                )

            # ------------------------------------------------
            # Y-axis limits
            # ------------------------------------------------

            ymin, ymax = y_axis_limits.get(metric, (None, None))

            if ymin is not None or ymax is not None:
                ax.set_ylim(ymin, ymax)

            # ------------------------------------------------
            # X-axis ticks and fixed x-limits
            # ------------------------------------------------

            xmax = df[time_col].max()
            ax.set_xlim(0, xmax)
            ax.set_xticks(np.arange(0, xmax + 30, 30))

            # ------------------------------------------------
            # Formatting
            # ------------------------------------------------

            ax.grid(True, which="both", linestyle="--", linewidth=0.5)
            ax.tick_params(axis="both", labelsize=AXIS_TICK_SIZE)

    # Do not use bbox_inches="tight"; keeps both images the same width.
    plt.savefig(output_file, dpi=300)
    plt.close()


# ============================================================
# Run plotting
# ============================================================

plot_teg_3x3(image_1_metrics, output_image_1)
plot_teg_3x3(image_2_metrics, output_image_2)
```

# CHEM 8 and CG4

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ==========================================================
# FILES
# ==========================================================

data_file = "CG4 and CHEM8 Data.xlsx"
tracker_file = "Experiment Tracker.xlsx"

output_dir = "CG4 and CHEM8 Plots"
os.makedirs(output_dir, exist_ok=True)

# ==========================================================
# TIME POINT CONTROL  (NUMERIC)
# ==========================================================

order_of_time_points = [
    0, 30, 60, 85, 120, 180, 240
]

# Control tick spacing here
X_TICK_INTERVAL = 30

# ==========================================================
# VISUAL SETTINGS
# ==========================================================

AXIS_TICK_SIZE = 16
LABEL_SIZE = 18
LEGEND_SIZE = 14

color_palette = [
    'black', 'blue', 'red',
    'purple', 'orange',
    'cyan', 'magenta'
]

occlusion_label_map = {
    'No Occlusion': 'No Aortic Occlusion',
    'Partial Occlusion': 'p-REBOA',
    'Full Occlusion': 'f-REBOA'
}

occlusion_order = [
    'No Occlusion',
    'Partial Occlusion',
    'Full Occlusion'
]

# ==========================================================
# METRIC RENAME MAPPING  (EDIT HERE)
# ==========================================================

metric_name_map = {
    "ph": "pH (unitless)",
    "tco2_x": "TCO₂ (CG4) (mmol/L)",
    "po2": "pO₂ (mmHg)",
    "pco2": "pCO₂ (mmHg)",
    "be_ecf": "Base Excess (ECF) (mmol/L)",
    "hco3": "HCO₃⁻ (mmol/L)",
    "so2": "O₂ Saturation (%)",
    "lac": "Lactate (mmol/L)",
    "temp": "Temperature (°C)",
    "temp_ph": "Temperature Corrected pH (unitless)",
    "temp_po2": "Temp Corrected pO₂ (mmHg)",
    "temp_pco2": "Temp Corrected pCO₂ (mmHg)",
    "cl": "Chloride (mmol/L)",
    "k": "Potassium (mmol/L)",
    "na": "Sodium (mmol/L)",
    "bun": "BUN (mg/dL)",
    "ica": "Ionized Calcium (mmol/L)",
    "glu": "Glucose (mg/dL)",
    "tco2_y": "TCO₂ (CHEM8) (mmol/L)",
    "angap": "Anion Gap (mmol/L)",
    "hb": "Hemoglobin (g/dL)",
    "hct": "Hematocrit (%)",
    "crea": "Creatinine (mg/dL)"
}

# ==========================================================
# LOAD TRACKER DATA
# ==========================================================

tracker = pd.read_excel(
    tracker_file,
    sheet_name=['T30 categorization', 'T60 categorization']
)

group_10 = set(tracker['T30 categorization']['Group_10'].dropna())
group_20 = set(tracker['T30 categorization']['Group_20'].dropna())
group_30 = set(tracker['T30 categorization']['Group_30'].dropna())

full_occlusion = set(tracker['T60 categorization']['Full_Occlusion'].dropna())
partial_occlusion = set(tracker['T60 categorization']['Partial_Occlusion'].dropna())
no_occlusion = set(tracker['T60 categorization']['No_Occlusion'].dropna())

# ==========================================================
# LOAD ALL SUBJECT SHEETS
# ==========================================================

all_sheets = pd.read_excel(data_file, sheet_name=None)

combined_df = []

for sheet_name, sheet_df in all_sheets.items():

    if 'name' in sheet_df.columns:
        sheet_df = sheet_df.drop(columns=['name'])

    sheet_df['Subject'] = sheet_df['Subject'].astype(str)

    def assign_hemorrhage(subj):
        if subj in group_10:
            return 10
        elif subj in group_20:
            return 20
        elif subj in group_30:
            return 30
        else:
            return np.nan

    def assign_occlusion(subj):
        if subj in full_occlusion:
            return 'Full Occlusion'
        elif subj in partial_occlusion:
            return 'Partial Occlusion'
        elif subj in no_occlusion:
            return 'No Occlusion'
        else:
            return np.nan

    sheet_df['Hemorrhage Level'] = sheet_df['Subject'].apply(assign_hemorrhage)
    sheet_df['Occlusion Group'] = sheet_df['Subject'].apply(assign_occlusion)

    combined_df.append(sheet_df)

df = pd.concat(combined_df, ignore_index=True)

# Remove subjects not in tracker
df = df.dropna(subset=['Hemorrhage Level', 'Occlusion Group'])

# ==========================================================
# FILTER TIME POINTS (NUMERIC SAFE)
# ==========================================================

df['Time Point'] = pd.to_numeric(df['Time Point'], errors='coerce')
df = df[df['Time Point'].isin(order_of_time_points)]

df = df.sort_values('Time Point')

# ==========================================================
# IDENTIFY METRICS
# ==========================================================

exclude_cols = [
    'Subject',
    'Time Point',
    'Hemorrhage Level',
    'Occlusion Group'
]

numeric_columns = df.select_dtypes(include=['number']).columns
metrics = [col for col in numeric_columns if col not in exclude_cols]

# ==========================================================
# PLOTTING FUNCTION
# ==========================================================

def create_plots_for_metric(metric):

    hemorrhage_levels = [10, 20, 30]
    fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

    handles, labels = [], []

    for ax, hemorrhage_level in zip(axes, hemorrhage_levels):

        level_df = df[df['Hemorrhage Level'] == hemorrhage_level]

        if level_df.empty:
            ax.set_title(f"{hemorrhage_level}% Hemorrhage - No Data",
                         fontsize=LABEL_SIZE)
            continue

        mean_df = level_df.groupby(
            ['Time Point', 'Occlusion Group'],
            as_index=False
        )[metric].mean()

        std_df = level_df.groupby(
            ['Time Point', 'Occlusion Group'],
            as_index=False
        )[metric].std()

        for i, occlusion in enumerate(occlusion_order):

            if occlusion not in mean_df['Occlusion Group'].unique():
                continue

            mean_subset = mean_df[mean_df['Occlusion Group'] == occlusion]
            std_subset = std_df[std_df['Occlusion Group'] == occlusion]

            color = color_palette[i % len(color_palette)]
            legend_label = occlusion_label_map.get(occlusion, occlusion)

            x_vals = mean_subset['Time Point']

            line, = ax.plot(
                x_vals,
                mean_subset[metric],
                marker='o',
                markersize=8,
                linewidth=2,
                label=legend_label,
                color=color
            )

            ax.fill_between(
                x_vals,
                mean_subset[metric] - std_subset[metric],
                mean_subset[metric] + std_subset[metric],
                alpha=0.2,
                color=color
            )

            if legend_label not in labels:
                handles.append(line)
                labels.append(legend_label)

        ax.set_title(f"{hemorrhage_level}% Hemorrhage",
                     fontsize=LABEL_SIZE)

        ax.set_xlabel("Time (mins)", fontsize=LABEL_SIZE)

        ax.set_xticks(
            np.arange(
                min(order_of_time_points),
                max(order_of_time_points) + 1,
                X_TICK_INTERVAL
            )
        )

        # ax.grid(True, which='both', linestyle='--', linewidth=0.5)
        ax.tick_params(axis='both', labelsize=AXIS_TICK_SIZE)

    y_label = metric_name_map.get(metric, metric)
    axes[0].set_ylabel(y_label, fontsize=LABEL_SIZE)

    axes[-1].legend(
        handles,
        labels,
        title="Intervention Group",
        loc="center left",
        bbox_to_anchor=(1.02, 0.5),
        fontsize=LEGEND_SIZE,
        title_fontsize=LEGEND_SIZE
    )

    plt.tight_layout()
    plt.savefig(
        os.path.join(output_dir, f"{metric}.png"),
        dpi=300,
        bbox_inches="tight"
    )
    plt.close()

# ==========================================================
# RUN ALL METRICS
# ==========================================================

for metric in metrics:
    create_plots_for_metric(metric)
```

# TEG Data

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# Parameters
# ============================================================

file_path = 'ERNE_TEG_cleaned.xlsx'

AXIS_TICK_SIZE = 16
TITLE_SIZE = 20
LABEL_SIZE = 18
PANEL_LABEL_SIZE = 18

output_dir = 'TEG'
os.makedirs(output_dir, exist_ok=True)

# ============================================================
# Output files
# ============================================================

output_image_1 = os.path.join(output_dir, "TEG_Image_1_R_Angle_G_3x3.png")
output_image_2 = os.path.join(output_dir, "TEG_Image_2_K_MA_LY30_3x3.png")

# ============================================================
# TEG metric layout
# These control which metrics go into each image and their panel letters
# ============================================================

image_1_metrics = [
    ("A", "R_time"),
    ("C", "Angle"),
    ("E", "G")
]

image_2_metrics = [
    ("B", "K_time"),
    ("D", "MA"),
    ("F", "LY30")
]

# ============================================================
# Manual panel titles and y-axis labels
#
# Edit these freely.
# panel_title = text shown above each graph next to A/B/C/etc.
# y_label = text shown on the y-axis.
# ============================================================

manual_labels = {
    "R_time": {
        "panel_title": "Reaction Time",
        "y_label": "Reaction Time (min)"
    },
    "K_time": {
        "panel_title": "Kinetic Time",
        "y_label": "Kinetic Time (min)"
    },
    "Angle": {
        "panel_title": "Alpha Angle",
        "y_label": "Alpha Angle (degrees)"
    },
    "MA": {
        "panel_title": "Maximum Amplitude",
        "y_label": "Maximum Amplitude (mm)"
    },
    "G": {
        "panel_title": "Shear Modulus",
        "y_label": "Shear Modulus"
    },
    "LY30": {
        "panel_title": "LY30",
        "y_label": "LY30 (%)"
    }
}

# ============================================================
# Y-axis limits
# Set min/max here for each graph.
#
# Use None if you want matplotlib to choose automatically.
# ============================================================

y_axis_limits = {
    "R_time": (None, None),
    "K_time": (None, None),

    # Based on feedback
    "Angle": (20, 85),
    "MA": (50, 85),
    "G": (5000, 18000),

    "LY30": (None, None)
}

# ============================================================
# Optional dotted vertical phase lines
#
# Edit these values based on your experimental phases.
# These will draw vertical dotted lines on every subplot.
# Set phase_lines = [] if you do not want them.
# ============================================================

phase_lines = [30, 60, 120, 180]

PHASE_LINE_COLOR = "gray"
PHASE_LINE_STYLE = ":"
PHASE_LINE_WIDTH = 1.5

# ============================================================
# Marker symbols and colors
# ============================================================

marker_info = {
    "No": {"marker": "o", "color": "black"},
    "Partial": {"marker": "o", "color": "blue"},
    "Full": {"marker": "o", "color": "red"}
}

legend_order = ["No", "Partial", "Full"]

# ============================================================
# Load and preprocess data
# ============================================================

df = pd.read_excel(file_path)

time_col = 'Time Point'
group_col = 'Occlusion Group'
level_col = 'Hemorrhage Level'

# Ensure numeric time
df[time_col] = pd.to_numeric(df[time_col], errors='coerce')
df = df.dropna(subset=[time_col])
df[time_col] = df[time_col].astype(float)

# Ensure numeric hemorrhage level
df[level_col] = pd.to_numeric(df[level_col], errors='coerce')

# Correct typo
df[group_col] = df[group_col].replace({
    "None Occlusion": "No Occlusion"
})

# Map occlusion group names to simplified plotting keys
def map_group_key(name):
    name = str(name)

    if "No" in name:
        return "No"
    elif "Partial" in name:
        return "Partial"
    elif "Full" in name:
        return "Full"

    return "No"

df["Group Key"] = df[group_col].apply(map_group_key)

# Make hemorrhage order explicit
hemorrhage_order = [10, 20, 30]

# ============================================================
# Plotting function
# ============================================================

def plot_teg_3x3(metric_layout, output_file):
    n_rows = len(metric_layout)
    n_cols = len(hemorrhage_order)

    fig, axes = plt.subplots(
        n_rows,
        n_cols,
        figsize=(18, 16),
        sharex=False,
        sharey=False
    )

    # Fixed margins keep both saved images and graph widths identical.
    # Increased top and hspace so labels can sit above graphs.
    fig.subplots_adjust(
        left=0.10,
        right=0.98,
        top=0.88,
        bottom=0.08,
        wspace=0.28,
        hspace=0.62
    )

    for row_idx, (panel_letter, metric) in enumerate(metric_layout):
        for col_idx, hemorrhage_level in enumerate(hemorrhage_order):

            ax = axes[row_idx, col_idx]

            level_df = df[df[level_col] == hemorrhage_level]

            grouped_df = (
                level_df
                .groupby(["Group Key", time_col])[metric]
                .agg(["mean", "std"])
                .reset_index()
            )

            for key in legend_order:
                subset = grouped_df[grouped_df["Group Key"] == key]

                if subset.empty:
                    continue

                ax.plot(
                    subset[time_col],
                    subset["mean"],
                    marker=marker_info[key]["marker"],
                    markersize=8,
                    linewidth=2,
                    color=marker_info[key]["color"]
                )

                ax.fill_between(
                    subset[time_col],
                    subset["mean"] - subset["std"],
                    subset["mean"] + subset["std"],
                    alpha=0.2,
                    color=marker_info[key]["color"]
                )

            # ------------------------------------------------
            # Dotted vertical phase demarcation lines
            # ------------------------------------------------

            for phase_time in phase_lines:
                ax.axvline(
                    x=phase_time,
                    color=PHASE_LINE_COLOR,
                    linestyle=PHASE_LINE_STYLE,
                    linewidth=PHASE_LINE_WIDTH
                )

            # ------------------------------------------------
            # Column titles only on top row, pushed higher
            # ------------------------------------------------

            if row_idx == 0:
                ax.set_title(
                    f"{hemorrhage_level}% Hemorrhage",
                    fontsize=TITLE_SIZE,
                    pad=44
                )

            # ------------------------------------------------
            # Manual panel labels above each graph
            # ------------------------------------------------

            panel_title = manual_labels[metric]["panel_title"]

            ax.text(
                0.00,
                1.10,
                f"{panel_letter}. {panel_title}",
                transform=ax.transAxes,
                fontsize=PANEL_LABEL_SIZE,
                fontweight="bold",
                va="bottom",
                ha="left",
                clip_on=False
            )

            # ------------------------------------------------
            # Axis labels
            # ------------------------------------------------

            ax.set_xlabel("Time", fontsize=LABEL_SIZE)

            if col_idx == 0:
                ax.set_ylabel(
                    manual_labels[metric]["y_label"],
                    fontsize=LABEL_SIZE
                )

            # ------------------------------------------------
            # Y-axis limits
            # ------------------------------------------------

            ymin, ymax = y_axis_limits.get(metric, (None, None))

            if ymin is not None or ymax is not None:
                ax.set_ylim(ymin, ymax)

            # ------------------------------------------------
            # X-axis ticks and fixed x-limits
            # ------------------------------------------------

            xmax = df[time_col].max()
            ax.set_xlim(0, xmax)
            ax.set_xticks(np.arange(0, xmax + 30, 30))

            # ------------------------------------------------
            # Formatting
            # ------------------------------------------------

            ax.grid(True, which="both", linestyle="--", linewidth=0.5)
            ax.tick_params(axis="both", labelsize=AXIS_TICK_SIZE)

    # Do not use bbox_inches="tight"; keeps both images the same width.
    plt.savefig(output_file, dpi=300)
    plt.close()


# ============================================================
# Run plotting
# ============================================================

plot_teg_3x3(image_1_metrics, output_image_1)
plot_teg_3x3(image_2_metrics, output_image_2)
```

# Supplemental PV Data

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re

# ------------------------
# 🔹 GLOBAL PLOT CONFIG
# ------------------------
PLOT_CONFIG = {
    "figsize": (40, 8),   # base size (we scale it dynamically)
    "dpi": 300,

    "axis_tick_size": 18,
    "label_size": 20,
    "legend_size": 16,

    "x_tick_interval": 15,
    "x_min": 0,
    "x_max": 240,
    # "x_max": 90,

    "line_width": 2,
    "marker_size": 8,
    "fill_alpha": 0.2,

    "ref_line_y": 1,
}

# ------------------------
# Terms to plot (ROWS)
# Use exact Excel column names here
# ------------------------
terms_to_plot = [
    "Ejection Fraction",
    "Stroke Volume",
    "Stroke Work",
    "Elastance"   # Excel column name; displayed below as Arterial Elastance
]

# Display names for row labels
row_label_map = {
    "Ejection Fraction": "Ejection Fraction",
    "Stroke Volume": "Stroke Volume",
    "Stroke Work": "Stroke Work",
    "Elastance": "Arterial Elastance"
}

# ------------------------
# Style definitions
# ------------------------
marker_info = {
    "None":    {"marker": "o", "color": "black"},
    "Partial": {"marker": "o", "color": "blue"},
    "Full":    {"marker": "o", "color": "red"}
}

legend_order = ["None", "Partial", "Full"]

occlusion_label_map = {
    "None": "No Aortic Occlusion",
    "No": "No Aortic Occlusion",
    "Partial": "p-REBOA",
    "Full": "f-REBOA"
}

# ------------------------
# Load and preprocess data
# ------------------------
file_path = 'OG-PV_Fractional_Increase_T0-1.xlsx'
df = pd.read_excel(file_path)

df = df.drop(columns=["Weight (kg)", "Gender"])

# 🔹 REQUIRED: define your time_point_order
# Example (replace with your real one if different)
time_point_order = df['Time Point'].astype(str).unique().tolist()

df['Time Point Raw'] = df['Time Point'].astype(str)
df = df[df['Time Point Raw'].isin(time_point_order)]

# ------------------------
# Time extraction
# ------------------------
def extract_time_point_number(time_point):
    match = re.match(r"T(\d+)-(\d+)", str(time_point))
    if match:
        first = int(match.group(1))
        second = int(match.group(2))

        if time_point in ["T29-30", "T239-240"]:
            return second

        return first
    return None

df['Time Point'] = df['Time Point Raw'].apply(extract_time_point_number)

# ------------------------
# Treatment parsing
# ------------------------
def extract_treatment_details(treatment):
    match = re.match(r"(\d+)%\s*(\w+)", str(treatment))
    if match:
        return match.group(1), match.group(2)
    return None, None

df[['Treatment Percentage', 'Treatment Group']] = df['Treatment Group'].apply(
    lambda x: pd.Series(extract_treatment_details(x))
)

df = df.sort_values('Time Point')

# 🔹 Hemorrhage levels (COLUMNS)
hemorrhage_levels = sorted(df['Hemorrhage Level'].dropna().unique())

# ------------------------
# Output folder
# ------------------------
output_folder = "Pressure-Volume"
os.makedirs(output_folder, exist_ok=True)

# ------------------------
# Plotting function
# ------------------------
def create_multiterm_panel():

    numeric_columns = df.select_dtypes(include=['number']).columns
    valid_terms = [t for t in terms_to_plot if t in numeric_columns]

    if len(valid_terms) == 0:
        raise ValueError(
            "None of the requested terms were found in the numeric Excel columns.\n"
            f"Requested terms: {terms_to_plot}\n"
            f"Available numeric columns: {list(numeric_columns)}"
        )

    n_rows = len(valid_terms)
    n_cols = len(hemorrhage_levels)

    fig, axes = plt.subplots(
        n_rows,
        n_cols,
        figsize=(10 * n_cols, 5 * n_rows),
        sharex=True,
        sharey='row'
    )

    if n_rows == 1:
        axes = np.array([axes])

    handles, labels = [], []

    for row_idx, term in enumerate(valid_terms):
        for col_idx, hemorrhage_level in enumerate(hemorrhage_levels):

            ax = axes[row_idx, col_idx]

            subset_df = df[df['Hemorrhage Level'] == hemorrhage_level]

            if subset_df.empty:
                ax.set_title(f"{float(hemorrhage_level):g}% - No Data")
                continue

            numeric_columns = subset_df.select_dtypes(include=['number']).columns

            mean_df = subset_df.groupby(
                ['Treatment Group', 'Time Point'],
                as_index=False
            )[numeric_columns].mean()

            std_df = subset_df.groupby(
                ['Treatment Group', 'Time Point'],
                as_index=False
            )[numeric_columns].std()

            mean_df['Time Point'] = mean_df['Time Point'].astype(int)
            std_df['Time Point'] = std_df['Time Point'].astype(int)

            for treatment_key in legend_order:
                mean_subset = mean_df[mean_df['Treatment Group'] == treatment_key]
                std_subset = std_df[std_df['Treatment Group'] == treatment_key]

                if mean_subset.empty:
                    continue

                color = marker_info[treatment_key]["color"]
                marker = marker_info[treatment_key]["marker"]

                legend_label = occlusion_label_map.get(treatment_key, treatment_key)

                line, = ax.plot(
                    mean_subset['Time Point'],
                    mean_subset[term],
                    marker=marker,
                    markersize=PLOT_CONFIG["marker_size"],
                    linewidth=PLOT_CONFIG["line_width"],
                    color=color,
                    label=legend_label
                )

                ax.fill_between(
                    mean_subset['Time Point'],
                    mean_subset[term] - std_subset[term],
                    mean_subset[term] + std_subset[term],
                    alpha=PLOT_CONFIG["fill_alpha"],
                    color=color
                )

                if legend_label not in labels:
                    handles.append(line)
                    labels.append(legend_label)

            # Reference line
            ax.axhline(
                y=PLOT_CONFIG["ref_line_y"],
                linestyle='--',
                linewidth=1
            )

            # Column titles: 10%, 20%, 30% instead of 10.0%, 20.0%, 30.0%
            if row_idx == 0:
                ax.set_title(
                    f"{float(hemorrhage_level):g}% Hemorrhage",
                    fontsize=PLOT_CONFIG["label_size"]
                )

            # Row labels
            if col_idx == 0:
                row_prefix = ["(A)", "(B)", "(C)", "(D)"][row_idx]
                display_term = row_label_map.get(term, term)

                ax.set_ylabel(
                    f"{row_prefix} {display_term}",
                    fontsize=PLOT_CONFIG["label_size"]
                )

            # Bottom x-labels
            if row_idx == n_rows - 1:
                ax.set_xlabel("Time (mins)", fontsize=PLOT_CONFIG["label_size"])

            # X ticks
            if PLOT_CONFIG["x_min"] is not None and PLOT_CONFIG["x_max"] is not None:
                start_tick = PLOT_CONFIG["x_min"]
                end_tick = PLOT_CONFIG["x_max"]
            else:
                min_time = mean_df['Time Point'].min()
                max_time = mean_df['Time Point'].max()
                start_tick = (min_time // PLOT_CONFIG["x_tick_interval"]) * PLOT_CONFIG["x_tick_interval"]
                end_tick = ((max_time // PLOT_CONFIG["x_tick_interval"]) + 1) * PLOT_CONFIG["x_tick_interval"]

            ax.set_xticks(
                np.arange(
                    start_tick,
                    end_tick + 1,
                    PLOT_CONFIG["x_tick_interval"]
                )
            )

            ax.tick_params(axis='both', labelsize=PLOT_CONFIG["axis_tick_size"])
            ax.grid(True, linestyle='--', linewidth=0.7)

    # # Shared legend
    # fig.legend(
    #     handles,
    #     labels,
    #     loc='center right',
    #     fontsize=PLOT_CONFIG["legend_size"]
    # )

    plt.tight_layout(rect=[0, 0, 1, 0.96])     # use FULL width

    filename = os.path.join(output_folder, "Supplemental Multiterm_Panel.png")
    plt.savefig(filename, dpi=PLOT_CONFIG["dpi"])
    plt.close()

# ------------------------
# Run
# ------------------------
create_multiterm_panel()
```

# Supplemental CG4 and CHEM8 Data

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ==========================================================
# FILES
# ==========================================================

data_file = "CG4 and CHEM8 Data.xlsx"
tracker_file = "Experiment Tracker.xlsx"

output_dir = "CG4 and CHEM8 Plots"
os.makedirs(output_dir, exist_ok=True)

# ==========================================================
# GLOBAL PLOT CONFIG
# ==========================================================

PLOT_CONFIG = {
    "figsize": (40, 30),
    "dpi": 300,

    "axis_tick_size": 18,
    "label_size": 20,
    "title_size": 28,
    "legend_size": 16,

    "x_tick_interval": 30,
    "x_min": 0,
    "x_max": 240,

    "line_width": 2,
    "marker_size": 8,
    "fill_alpha": 0.2,
}

# Optional: make matplotlib text bold by default
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["axes.titleweight"] = "bold"

# ==========================================================
# TIME POINT CONTROL
# ==========================================================

order_of_time_points = [
    0, 30, 60, 85, 120, 180, 240
]

# ==========================================================
# METRICS TO PLOT AS PANEL ROWS
# Use exact Excel column names here
# ==========================================================

metrics_to_plot = [
    "k",
    "glu",
    "ph",
    "po2",
    "pco2",
    "hco3"
]

row_label_map = {
    "k": "(A) Potassium",
    "glu": "(B) Glucose",
    "ph": "(C) pH",
    "po2": "(D) pO₂",
    "pco2": "(E) pCO₂",
    "hco3": "(F) Bicarbonate"
}

# ==========================================================
# STYLE DEFINITIONS
# ==========================================================

marker_info = {
    "No Occlusion":      {"marker": "o", "color": "black"},
    "Partial Occlusion": {"marker": "o", "color": "blue"},
    "Full Occlusion":    {"marker": "o", "color": "red"}
}

occlusion_order = [
    "No Occlusion",
    "Partial Occlusion",
    "Full Occlusion"
]

occlusion_label_map = {
    "No Occlusion": "No Aortic Occlusion",
    "Partial Occlusion": "p-REBOA",
    "Full Occlusion": "f-REBOA"
}

# ==========================================================
# LOAD TRACKER DATA
# ==========================================================

tracker = pd.read_excel(
    tracker_file,
    sheet_name=["T30 categorization", "T60 categorization"]
)

group_10 = set(tracker["T30 categorization"]["Group_10"].dropna().astype(str))
group_20 = set(tracker["T30 categorization"]["Group_20"].dropna().astype(str))
group_30 = set(tracker["T30 categorization"]["Group_30"].dropna().astype(str))

full_occlusion = set(tracker["T60 categorization"]["Full_Occlusion"].dropna().astype(str))
partial_occlusion = set(tracker["T60 categorization"]["Partial_Occlusion"].dropna().astype(str))
no_occlusion = set(tracker["T60 categorization"]["No_Occlusion"].dropna().astype(str))

# ==========================================================
# LOAD ALL SUBJECT SHEETS
# ==========================================================

all_sheets = pd.read_excel(data_file, sheet_name=None)

combined_df = []

for sheet_name, sheet_df in all_sheets.items():

    if "name" in sheet_df.columns:
        sheet_df = sheet_df.drop(columns=["name"])

    sheet_df["Subject"] = sheet_df["Subject"].astype(str)

    def assign_hemorrhage(subj):
        if subj in group_10:
            return 10
        elif subj in group_20:
            return 20
        elif subj in group_30:
            return 30
        else:
            return np.nan

    def assign_occlusion(subj):
        if subj in full_occlusion:
            return "Full Occlusion"
        elif subj in partial_occlusion:
            return "Partial Occlusion"
        elif subj in no_occlusion:
            return "No Occlusion"
        else:
            return np.nan

    sheet_df["Hemorrhage Level"] = sheet_df["Subject"].apply(assign_hemorrhage)
    sheet_df["Occlusion Group"] = sheet_df["Subject"].apply(assign_occlusion)

    combined_df.append(sheet_df)

df = pd.concat(combined_df, ignore_index=True)

# Remove subjects not in tracker
df = df.dropna(subset=["Hemorrhage Level", "Occlusion Group"])

# ==========================================================
# FILTER TIME POINTS
# ==========================================================

df["Time Point"] = pd.to_numeric(df["Time Point"], errors="coerce")
df = df[df["Time Point"].isin(order_of_time_points)]

df = df.sort_values("Time Point")

# ==========================================================
# MAKE SURE REQUESTED METRICS EXIST
# ==========================================================

numeric_columns = df.select_dtypes(include=["number"]).columns
valid_metrics = [m for m in metrics_to_plot if m in numeric_columns]

if len(valid_metrics) == 0:
    raise ValueError(
        "None of the requested metrics were found in the numeric Excel columns.\n"
        f"Requested metrics: {metrics_to_plot}\n"
        f"Available numeric columns: {list(numeric_columns)}"
    )

missing_metrics = [m for m in metrics_to_plot if m not in numeric_columns]

if len(missing_metrics) > 0:
    print("Warning: These requested metrics were not found and will be skipped:")
    print(missing_metrics)
    print("Available numeric columns:")
    print(list(numeric_columns))

# ==========================================================
# PANEL PLOTTING FUNCTION
# ==========================================================

def create_multimetric_panel():

    hemorrhage_levels = [10, 20, 30]

    n_rows = len(valid_metrics)
    n_cols = len(hemorrhage_levels)

    fig, axes = plt.subplots(
        n_rows,
        n_cols,
        figsize=(10 * n_cols, 5 * n_rows),
        sharex=True,
        sharey="row"
    )

    if n_rows == 1:
        axes = np.array([axes])

    handles, labels = [], []

    for row_idx, metric in enumerate(valid_metrics):
        for col_idx, hemorrhage_level in enumerate(hemorrhage_levels):

            ax = axes[row_idx, col_idx]

            level_df = df[df["Hemorrhage Level"] == hemorrhage_level]

            if level_df.empty:
                ax.set_title(
                    f"{hemorrhage_level:g}% Hemorrhage - No Data",
                    fontsize=PLOT_CONFIG["title_size"],
                    fontweight="bold"
                )
                continue

            mean_df = level_df.groupby(
                ["Occlusion Group", "Time Point"],
                as_index=False
            )[metric].mean()

            std_df = level_df.groupby(
                ["Occlusion Group", "Time Point"],
                as_index=False
            )[metric].std()

            mean_df["Time Point"] = mean_df["Time Point"].astype(int)
            std_df["Time Point"] = std_df["Time Point"].astype(int)

            for occlusion_key in occlusion_order:

                mean_subset = mean_df[mean_df["Occlusion Group"] == occlusion_key]
                std_subset = std_df[std_df["Occlusion Group"] == occlusion_key]

                if mean_subset.empty:
                    continue

                mean_subset = mean_subset.sort_values("Time Point")
                std_subset = std_subset.sort_values("Time Point")

                color = marker_info[occlusion_key]["color"]
                marker = marker_info[occlusion_key]["marker"]

                legend_label = occlusion_label_map.get(occlusion_key, occlusion_key)

                line, = ax.plot(
                    mean_subset["Time Point"],
                    mean_subset[metric],
                    marker=marker,
                    markersize=PLOT_CONFIG["marker_size"],
                    linewidth=PLOT_CONFIG["line_width"],
                    color=color,
                    label=legend_label
                )

                ax.fill_between(
                    mean_subset["Time Point"],
                    mean_subset[metric] - std_subset[metric],
                    mean_subset[metric] + std_subset[metric],
                    alpha=PLOT_CONFIG["fill_alpha"],
                    color=color
                )

                if legend_label not in labels:
                    handles.append(line)
                    labels.append(legend_label)

            # Column titles
            if row_idx == 0:
                ax.set_title(
                    f"{hemorrhage_level:g}% Hemorrhage",
                    fontsize=PLOT_CONFIG["title_size"],
                    fontweight="bold"
                )

            # Row labels
            if col_idx == 0:
                ax.set_ylabel(
                    row_label_map.get(metric, metric),
                    fontsize=PLOT_CONFIG["label_size"],
                    fontweight="bold"
                )

            # Bottom x-labels
            if row_idx == n_rows - 1:
                ax.set_xlabel(
                    "Time (mins)",
                    fontsize=PLOT_CONFIG["label_size"],
                    fontweight="bold"
                )

            # X ticks
            ax.set_xticks(
                np.arange(
                    PLOT_CONFIG["x_min"],
                    PLOT_CONFIG["x_max"] + 1,
                    PLOT_CONFIG["x_tick_interval"]
                )
            )

            ax.set_xlim(
                PLOT_CONFIG["x_min"],
                PLOT_CONFIG["x_max"]
            )

            ax.tick_params(
                axis="both",
                labelsize=PLOT_CONFIG["axis_tick_size"]
            )

            # Bold tick labels
            for tick in ax.get_xticklabels() + ax.get_yticklabels():
                tick.set_fontweight("bold")

            ax.grid(True, linestyle="--", linewidth=0.7)

    # Legend intentionally not plotted
    # fig.legend(
    #     handles,
    #     labels,
    #     loc="center right",
    #     fontsize=PLOT_CONFIG["legend_size"]
    # )

    plt.tight_layout(rect=[0, 0, 1, 0.96])

    filename = os.path.join(
        output_dir,
        "CG4_CHEM8_Multimetric_Panel.png"
    )

    plt.savefig(
        filename,
        dpi=PLOT_CONFIG["dpi"]
    )

    plt.close()

# ==========================================================
# RUN
# ==========================================================

create_multimetric_panel()
```

# REMADE TEG DATA

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# FILES
# ============================================================

file_path = "ERNE_TEG_cleaned.xlsx"

output_dir = "TEG"
os.makedirs(output_dir, exist_ok=True)

# ============================================================
# GLOBAL PLOT CONFIG
# ============================================================

PLOT_CONFIG = {
    "figsize": (40, 30),
    "dpi": 300,

    "axis_tick_size": 18,
    "label_size": 20,
    "title_size": 28,
    "legend_size": 16,

    "x_tick_interval": 30,
    "x_min": 0,
    "x_max": 240,

    "line_width": 2,
    "marker_size": 8,
    "fill_alpha": 0.2,
}

# Optional: make matplotlib text bold by default
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["axes.titleweight"] = "bold"

# ============================================================
# OUTPUT FILE
# ============================================================

output_image = os.path.join(output_dir, "TEG_Multimetric_Panel.png")

# ============================================================
# TEG METRICS TO PLOT AS PANEL ROWS
# Use exact Excel column names here
# ============================================================

metrics_to_plot = [
    "R_time",
    "K_time",
    "Angle",
    "MA",
    "G",
    "LY30"
]

row_label_map = {
    "R_time": "(A) Reaction Time",
    "K_time": "(B) Kinetic Time",
    "Angle": "(C) Alpha Angle",
    "MA": "(D) Maximum Amplitude",
    "G": "(E) Shear Modulus",
    "LY30": "(F) LY30"
}

# ============================================================
# Y-AXIS LABELS
# ============================================================

y_label_map = {
    "R_time": "Reaction Time (min)",
    "K_time": "Kinetic Time (min)",
    "Angle": "Alpha Angle (degrees)",
    "MA": "Maximum Amplitude (mm)",
    "G": "Shear Modulus",
    "LY30": "LY30 (%)"
}

# ============================================================
# Y-AXIS LIMITS
# Use None if you want matplotlib to choose automatically.
# ============================================================

y_axis_limits = {
    "R_time": (None, None),
    "K_time": (None, None),
    "Angle": (20, 85),
    "MA": (50, 85),
    "G": (5000, 18000),
    "LY30": (None, None)
}

# ============================================================
# OPTIONAL DOTTED VERTICAL PHASE LINES
# Set phase_lines = [] if you do not want them.
# ============================================================

phase_lines = [30, 60, 120, 180]

PHASE_LINE_COLOR = "gray"
PHASE_LINE_STYLE = ":"
PHASE_LINE_WIDTH = 1.5

# ============================================================
# STYLE DEFINITIONS
# ============================================================

marker_info = {
    "No":      {"marker": "o", "color": "black"},
    "Partial": {"marker": "o", "color": "blue"},
    "Full":    {"marker": "o", "color": "red"}
}

legend_order = [
    "No",
    "Partial",
    "Full"
]

occlusion_label_map = {
    "No": "No Aortic Occlusion",
    "Partial": "p-REBOA",
    "Full": "f-REBOA"
}

# ============================================================
# LOAD AND PREPROCESS DATA
# ============================================================

df = pd.read_excel(file_path)

time_col = "Time Point"
group_col = "Occlusion Group"
level_col = "Hemorrhage Level"

# Ensure numeric time
df[time_col] = pd.to_numeric(df[time_col], errors="coerce")
df = df.dropna(subset=[time_col])
df[time_col] = df[time_col].astype(float)

# Ensure numeric hemorrhage level
df[level_col] = pd.to_numeric(df[level_col], errors="coerce")

# Correct typo
df[group_col] = df[group_col].replace({
    "None Occlusion": "No Occlusion"
})

# Map occlusion group names to simplified plotting keys
def map_group_key(name):
    name = str(name)

    if "No" in name:
        return "No"
    elif "Partial" in name:
        return "Partial"
    elif "Full" in name:
        return "Full"

    return "No"

df["Group Key"] = df[group_col].apply(map_group_key)

# Hemorrhage columns
hemorrhage_order = [10, 20, 30]

# ============================================================
# MAKE SURE REQUESTED METRICS EXIST
# ============================================================

numeric_columns = df.select_dtypes(include=["number"]).columns
valid_metrics = [m for m in metrics_to_plot if m in numeric_columns]

if len(valid_metrics) == 0:
    raise ValueError(
        "None of the requested TEG metrics were found in the numeric Excel columns.\n"
        f"Requested metrics: {metrics_to_plot}\n"
        f"Available numeric columns: {list(numeric_columns)}"
    )

missing_metrics = [m for m in metrics_to_plot if m not in numeric_columns]

if len(missing_metrics) > 0:
    print("Warning: These requested TEG metrics were not found and will be skipped:")
    print(missing_metrics)
    print("Available numeric columns:")
    print(list(numeric_columns))

# ============================================================
# PANEL PLOTTING FUNCTION
# ============================================================

def create_teg_multimetric_panel():

    n_rows = len(valid_metrics)
    n_cols = len(hemorrhage_order)

    fig, axes = plt.subplots(
        n_rows,
        n_cols,
        figsize=(10 * n_cols, 5 * n_rows),
        sharex=True,
        sharey=False
    )

    if n_rows == 1:
        axes = np.array([axes])

    handles, labels = [], []

    for row_idx, metric in enumerate(valid_metrics):
        for col_idx, hemorrhage_level in enumerate(hemorrhage_order):

            ax = axes[row_idx, col_idx]

            level_df = df[df[level_col] == hemorrhage_level]

            if level_df.empty:
                ax.set_title(
                    f"{hemorrhage_level:g}% Hemorrhage - No Data",
                    fontsize=PLOT_CONFIG["title_size"],
                    fontweight="bold"
                )
                continue

            grouped_df = (
                level_df
                .groupby(["Group Key", time_col])[metric]
                .agg(["mean", "std"])
                .reset_index()
            )

            for group_key in legend_order:

                subset = grouped_df[grouped_df["Group Key"] == group_key]

                if subset.empty:
                    continue

                subset = subset.sort_values(time_col)

                color = marker_info[group_key]["color"]
                marker = marker_info[group_key]["marker"]

                legend_label = occlusion_label_map.get(group_key, group_key)

                line, = ax.plot(
                    subset[time_col],
                    subset["mean"],
                    marker=marker,
                    markersize=PLOT_CONFIG["marker_size"],
                    linewidth=PLOT_CONFIG["line_width"],
                    color=color,
                    label=legend_label
                )

                ax.fill_between(
                    subset[time_col],
                    subset["mean"] - subset["std"],
                    subset["mean"] + subset["std"],
                    alpha=PLOT_CONFIG["fill_alpha"],
                    color=color
                )

                if legend_label not in labels:
                    handles.append(line)
                    labels.append(legend_label)

            # Dotted vertical phase demarcation lines
            for phase_time in phase_lines:
                ax.axvline(
                    x=phase_time,
                    color=PHASE_LINE_COLOR,
                    linestyle=PHASE_LINE_STYLE,
                    linewidth=PHASE_LINE_WIDTH
                )

            # Column titles
            if row_idx == 0:
                ax.set_title(
                    f"{hemorrhage_level:g}% Hemorrhage",
                    fontsize=PLOT_CONFIG["title_size"],
                    fontweight="bold"
                )

            # Row labels
            if col_idx == 0:
                ax.set_ylabel(
                    row_label_map.get(metric, metric),
                    fontsize=PLOT_CONFIG["label_size"],
                    fontweight="bold"
                )

            # Bottom x-labels
            if row_idx == n_rows - 1:
                ax.set_xlabel(
                    "Time (mins)",
                    fontsize=PLOT_CONFIG["label_size"],
                    fontweight="bold"
                )

            # Y-axis limits
            ymin, ymax = y_axis_limits.get(metric, (None, None))

            if ymin is not None or ymax is not None:
                ax.set_ylim(ymin, ymax)

            # X ticks
            ax.set_xticks(
                np.arange(
                    PLOT_CONFIG["x_min"],
                    PLOT_CONFIG["x_max"] + 1,
                    PLOT_CONFIG["x_tick_interval"]
                )
            )

            ax.set_xlim(
                PLOT_CONFIG["x_min"],
                PLOT_CONFIG["x_max"]
            )

            ax.tick_params(
                axis="both",
                labelsize=PLOT_CONFIG["axis_tick_size"]
            )

            # Bold tick labels
            for tick in ax.get_xticklabels() + ax.get_yticklabels():
                tick.set_fontweight("bold")

            ax.grid(True, linestyle="--", linewidth=0.7)

    # Legend intentionally not plotted
    # fig.legend(
    #     handles,
    #     labels,
    #     loc="center right",
    #     fontsize=PLOT_CONFIG["legend_size"]
    # )

    plt.tight_layout(rect=[0, 0, 1, 0.96])

    plt.savefig(
        output_image,
        dpi=PLOT_CONFIG["dpi"]
    )

    plt.close()

# ============================================================
# RUN
# ============================================================

create_teg_multimetric_panel()
```
